# Creating Target Variables

```python
#Finding the median absenteeism hours to set classification threshold
data_preprocessed["Absenteeism Time in Hours"].median()
```

```python
#Create the targets variable using the median value as a cut-off point. Values of 1 represent excessive absenteeism while
#values of 0 represent "normal" levels of absenteeism. 
targets = np.where(data_preprocessed["Absenteeism Time in Hours"] > 
                   data_preprocessed["Absenteeism Time in Hours"].median(), 1, 0)
```

```python
data_preprocessed["Excessive Absenteeism"] = targets
```

```python
data_with_targets = data_preprocessed.drop(["Absenteeism Time in Hours", 'Day of the Week', 'Distance to Work', 
                                            "Daily Work Load Average"], axis = 1)
```

# Selecting inputs for regression

```python
#Use the .iloc method to isolate the first 14 columns
data_with_targets.iloc[:, :-1].head()
```

```python
#Omit the last column "Excessive Absenteeism" as it is the target variable.
unscaled_inputs = data_with_targets.iloc[:, :-1]
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, columns, copy = True, with_mean = True, with_std = True):
        self.scaler = StandardScaler(copy, with_mean, with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y = None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y = None, copy = None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns = self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis = 1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
#columns_to_scale = ['Month Value','Day of the Week', 'Transportation Expense', 'Distance to Work',
       #'Age', 'Daily Work Load Average', 'Body Mass Index', 'Children', 'Pets']
    
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
                   
```

```python
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
```

```python
absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
#the .fit() method will calculate the mean and standard deviation of the input variables
#to prepare for scaling.
absenteeism_scaler.fit(unscaled_inputs)
```

```python
#Use the .transform method to scale the inputs and save it to a new variable scaled_inputs
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
#Verify the observations were scaled 
scaled_inputs.head()
```

```python
scaled_inputs.shape
```

# Splitting the data into Train and Test datasets

```python
#import the train_test_split module from scikit
from sklearn.model_selection import train_test_split
```

```python
#Use the input and target variable dataframes in the split method.
#Creates 4 separate arrays:
#Array 1 = Training dataset with inputs
#Array 2 = Training dataset with targets
#Array 3 = Test dataset with inputs
#Array 4 = Test dataset with targets
train_test_split(scaled_inputs, targets)
```

```python
#Save the output arrays into different variables.
#Use train_size to set the training data to an 80/20 split.
#Set random_state to 20 to make the split data pseudo-random.
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size = 0.8, random_state = 20)
```

```python
#Verify the split and examine the array shape
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```
